```python
def func(x):
    return (x+3)**2

def derivative(x):
    return 2*(x+3)
```

```python
gd=[]
def gradient_descent(starting_x,learning_rate,iterations):
    gd.append(starting_x)
    x=starting_x
    for i in range(iterations):
        deri=derivative(x)
        x=x-learning_rate*deri
        gd.append(x)
        print(f"Iteration {i+1}  x{i}={x} and f(x)={func(x)}")
        
    return x
```

```python
min_x=gradient_descent(5,0.1,50)
print(f"minima is at {min_x}")
```

```python
import seaborn as sns
sns.scatterplot(gd)
```
